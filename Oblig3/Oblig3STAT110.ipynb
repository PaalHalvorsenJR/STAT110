### Oppgave 1 a)

$ P(E) = 0.4, P(F) = 0.5, P(E \cup F) = 0.8 $

Disjunkte hendelser =  to hendelser er disjunkte dersom de ikke kan inntreffe samtidig.
Matematisk  er $ E \cup F $ disjunkt dersom $ P(E \cup F) = 0 $

$ P(E \cup F) = P(E) + P(F) - P(E \cap F) $

$  0.8 = 0.5 + 0.4 - P(E \cap F) $

$ P(E \cap F) = 0.1 $

$ Siden P(E \cap F) = 0.1 $ er $ E \cap F $ ikke disjunkt.

For å sjekke $E \cap F $ er uavhengig av hverandre må vi sjekke om $ P(E \cap F) = P(E) \cdot P(F) $

$ 0.1 = 0.4 \cdot 0.5 $

$ 0.1 = 0.2 $

Siden $ 0.1 \neq 0.2 $ er $ E \cap F $ ikke uavhengig av hverandre.



```python
# Part (a)
p_e = 0.4 
p_f = 0.5
p_e_union_f = 0.8 

p_e_intersect_f = p_e + p_f - p_e_union_f

print("P(E ∩ F) = ", p_e_intersect_f)

def is_depended_or_not(x,y,z):
    math = (x * y)
    if math == z:
        print("Hendelsene er uavhengi, ")
    else:
        print("Hendelsene er avhengi, siden vi fikk en ulikhet istedenfor en likhet ")

is_depended_or_not(p_e,p_f,p_e_intersect_f)


```

# Oljeleting

Vi har to oljefelt, C og D, med følgende hendelser:
- C = "oljefunn på felt 1"
- C' = "ingen olje på felt 1"
- D = "oljefunn på felt 2"
- D' = "ingen olje på felt 2" 

Vi har også fått oppgitt følgende sannsynligheter:
- $ P(C \cap D) = 0.07$
- $ P(C' \cap D) = 0.11 $
- $ P(C \cap D') = 0.14 $
- $ P(C' \cap  D') = 0.68$

#### Sannsynlighet for oljefunn på felt 1

Først må vi finne sannsynligheten for at C inntreffer

$ P(C) = P(C \cap D) + P(C \cap D') $

$ P(C) = 0.07 + 0.14 $

$ P(C) = 0.21 $

Sannsynligheten for oljefunn på felt 1 er 0.21 eller 21%

#### Sannsynligheten for oljefunn på felt 1 gitt at det også er olje på felt 2

Nå er det gitt at det også er olje på felt 2

$ P(C | D) = \frac{P(C \cap D)}{P(D)} $

$ P(C | D) = \frac{0.07}{P(D)} $

$ P(D) = P(C \cap D) + P(C' \cap D) $

$ P(D) = 0.07 + 0.11 $

$ P(C | D) = \frac{0.07}{0.18} $

$ P(C | D) = 0.3888888888888889 $

Sannsynligheten for at det er olje på felt 1 gitt at det er olje på felt 2 er 0.3888888888888889 eller 38.89%

#### Sannsynligheten for oljefunn på felt 1 gitt at det ikke er olje på felt 2

Nå er det gitt at det ikke er olje på felt 2
Først må vi finne sannsynligheten for den komplementære hendelsen til D

$ P(D') = P(C \cap D') + P(C' \cap D') $

$ P(D') = 0.14 + 0.68 $

$ P(D') = 0.82 $

$ P(C | D') = \frac{P(C \cap D')}{P(D')} $

$ P(C | D') = \frac{0.14}{0.82} $

$ P(C | D') = 0.17073170731707318 $

Sannsynligheten for at det er olje på felt 1 gitt at det ikke er olje på felt 2 er 0.17073170731707318 eller 17.07%

#### Uavhengighet mellom C og D 

For at to hendelser skal være uavhengig av hverandre må $ P(C \cap D) = P(C) \cdot P(D) $

Først må vi finne P(D)

$ P(D) = P(C \cap D) + P(C' \cap D) $

$ P(D) = 0.07 + 0.11 $

$ P(D) = 0.18 $

$ P(C) \cdot P(D) = 0.21 \cdot 0.18 $

$ P(C) \cdot P(D) = 0.0378 $

$ P(C \cap D) = 0.07 $

$ 0.07 \neq 0.0378 $

Siden $ P(C \cap D) \neq P(C) \cdot P(D) $ er C og D ikke uavhengig av hverandre.

```python
# Oppgave 1b

# Given probabilities
p_cd = 0.07
p_d_comp_c = 0.11
p_c_comp_d = 0.14
p_comp_c_comp_d = 0.68

def oljefelt1(x,y):
    math = (x + y)
    print("Sannsynligheten for at det er olje i felt 1 er: ", round(math,3 ))

def oljefelt1gittfelt2(x,y):
    math = x + y 
    p_c_d = x / math
    print("Sannsynligheten for at det er olje i felt 1 gitt at det er olje i felt 2 er: ", round(p_c_d,3))

def oljefelt1ikkegittfelt2(x,y):
    p_comp_d= x + y 
    p_c_comp_d = x / p_comp_d
    print("Sannsynligheten for at det er olje i felt 1 gitt at det ikke er olje i felt 2 er: ", round(p_c_comp_d,3))

def uavhengig_eller_avhengig(p_cd, p_c, p_d):
    if p_cd == p_c * p_d:
        print("Hendelsene er uavhengige, siden vi fikk en likhet istedenfor en ulikhet")
    else:
        print("Hendelsene er avhengige, siden vi fikk en ulikhet istedenfor en likhet")



oljefelt1(p_cd, p_c_comp_d)

oljefelt1gittfelt2(p_cd,p_d_comp_c)

oljefelt1ikkegittfelt2(p_c_comp_d,p_comp_c_comp_d)

#Beregn total sannsynlighet for C og D
p_c = p_cd + p_c_comp_d
p_d = p_cd + p_d_comp_c

uavhengig_eller_avhengig(p_cd, p_c, p_d)
```

For å beregne forventningen og variansen til U, kan vi bruke formlene for forventning og varians til en binomisk fordeling. for linære kombinasjoner av stokastiske variabler gjelder følgende:

La X og Y representere avkastningen til Orkla og Norsk Hydro, respektivt og la U = 0.3X + 0.7Y.

#### Forventning til U (E(U))

$ E(X) = 0.4 ,E(Y) = 0.6 $

$ U = qX + (1-q)Y $

$ U = 0.3X + (1-0.3)Y $

$ U = 0.3X + 0.7Y $

$ E(U) = E(0.3X + 0.7Y) $

$ E(U) = 0.3E(X) + 0.7E(Y) $

$ E(U) = 0.3 \cdot 0.4 + 0.7 \cdot 0.6 $

$ E(U) = 0.12 + 0.42 $

$ E(U) = 0.54 $

Forventningen til U er 0.54 eller 54%

### Varians til U(V(U))

$ V(x) = 0.25, V(Y) = 0.49, U = qX + (1-q)Y, q = 0.3, \rho (XY) = 0.3$, 

variansen av en linær kombinasjon av to korrrelerte stokastiske variabler er gitt ved:

$ V(U) = q^2V(X) + (1-q)^2V(Y) + 2(1-q) \cdot \rho (X,Y) \cdot \sqrt{V(X)V(Y)} $

$ \sigma V(X) = \sqrt{0.25} = 0.5 $

$ \sigma V(Y) = \sqrt{0.49} = 0.7  $

$ V(U) = 0.3^2 \cdot 0.25 + (1-0.3)^2 \cdot 0.49 + 2(0.3)(0.7)(0.3) \cdot \sqrt{0.25 * 0.49} $

$ V(U) = 0.0225 + 0.2401 + 0.0441 $

$ V(U) = 0.3067 $







```python
import math

# verdier 
E_X = 0.04
Var_X = 0.25
E_Y = 0.06
Var_Y = 0.49
rho_XY = 0.3
q = 0.3

# kalkulerer forventningsverdien til U
E_U = q * E_X + (1 - q) * E_Y
print("Forventningen til U:", round(E_U,3))

# kalukulerer variansen til U
Var_U = q**2 * Var_X + (1 - q)**2 * Var_Y + 2 * q * (1 - q) * rho_XY * math.sqrt(Var_X) * math.sqrt(Var_Y)
print("Variansen for U:", Var_U)
```

### Oppgave 3 a)

For å finne marginaltettheten fx(x) må vi summere over alle verdier av y (y = 0 og y = 1) produktet av den betingede tettheten fx|y(x|y) og sansynligheten for py(y). Utrykket er gitt ved:

$ f_x(x) = \sum_{y} f_{x|y}(x|y) \cdot P_y(y) $

$ f_x(x) = f_{x|y}(x|0) \cdot P_y(0) + f_{x|y}(x|1) \cdot P_y(1) $

vi vet at py (0) = py (1) = 0.5 dermed:

$ f_x(x) = \frac{3}{100}(1- \frac{x}{100}^2 ) + \frac{1}{2} + \frac{3}{100} (\frac{x}{100})^2 \cdot \frac{1}{2} $

$ f_x(x) = \frac{3}{100}[(1- \frac{x}{100}^2 + (\frac{x}{100})^2 )] $



```python
import numpy as np
import matplotlib.pyplot as plt


def f_X_gitt_Y(x, y):
    if y == 0:
        return 3/100 * (1 - x/100)**2
    else:
        return 3/100 * (x/100)**2


def p_Y(y):
    return 1/2


def f_X(x):
    return sum([f_X_gitt_Y(x, y) * p_Y(y) for y in [0, 1]])


x_values = np.linspace(0, 100, 1000)
y_values = [f_X(x) for x in x_values]
plt.plot(x_values, y_values)
plt.xlabel('x')
plt.ylabel('f_X(x)')
plt.title('Marginal Density f_X(x)')
plt.show()
```

#### a)


$ P(X = x) = \frac{e^{-\theta} \cdot \theta^x}{x!} $

$ P(Y = y|2\theta) = \frac{e^{-2\theta} \cdot (2\theta)^y}{y!} $

siden x og y er uahengie, er likelihood-funksjonen for de gitt observasjonene gitt ved produktet av de to sannsynlighetene:

$ L(\theta) = P(X = 2, Y = 5| \theta) = P(X = 2 | \theta) \cdot P(Y = 5|2 \theta) = \frac{e^{-\theta} \cdot \theta^2}{2!} \cdot \frac{e^{-2\theta} \cdot (2\theta)^5}{5!} $

nå kan vi forenkle utrykket ved å bruke begge egenskapene til logaritmer: og ved å sette inn de gitte verdiene for x = 2 og y = 5:

$ L(\theta) = -\theta + 2 \log (\theta)- \log (2!) - 2\theta + 5 \log (2\theta) - \log (5!)$

$ L(\theta) = -3\theta + 7 \log (\theta) + 5 \log (2\theta) - \log (2!) - \log (5!)$

#### b)

$ l(\theta) = [5 \ln - \ln(2!)] + 7 \ln \theta - 3 \theta $

dervivert av $l(\theta)$ med hensyn på \theta er:

$ \frac{dl(\theta)}{d\theta} = \frac{7}{\theta} - 3 $

setter vi dette lik 0 får vi:

$ \frac{7}{\theta} - 3 = 0 $

$ \frac{7}{\theta} = 3 $

$ 7 = 3\theta $

$ \theta = \frac{7}{3} $



```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Definer theta-verdier
theta = np.linspace(0.1, 10, 400)  # starter med 0.1 for å unngå dividere med null

# Beregn l(theta)
l_theta = 5*np.log(2) - math.log(math.factorial(2)) - math.log(math.factorial(5)) + 7*np.log(theta) - 3*theta

# Lag plott
plt.plot(theta, l_theta)
plt.xlabel('theta')
plt.ylabel('l(theta)')
plt.title('Log-Likelihood Function l(theta)')
plt.grid(True)
plt.axvline(x=theta[np.argmax(l_theta)], color='red', linestyle='--')  
plt.show()

# Print den theta-verdien der l(theta) er maksimum
print('Maksimum l(theta) er ved theta ≈', round(theta[np.argmax(l_theta)],2))
```
